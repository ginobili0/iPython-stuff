```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.factorplot('Sex', data = titanic_df, kind = 'count')
```

```python
sns.factorplot('Sex', data = titanic_df, hue = 'Pclass', kind = 'count')
```

```python
sns.factorplot('Pclass', data = titanic_df, hue = 'Sex', kind = 'count')
```

```python
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age', 'Sex']].apply(male_female_child, axis = 1)
```

```python
titanic_df.head(10)
```

```python
sns.factorplot('Pclass', data = titanic_df, kind = 'count', hue = 'person')
```

```python
titanic_df['Age'].mean()
```

```python
titanic_df['person'].value_counts()
```

```python
fig = sns.FacetGrid(titanic_df, hue = 'Sex', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()
fig.set(xlim = (0, oldest))

fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df, hue = 'person', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()
fig.set(xlim = (0, oldest))

fig.add_legend()
```

```python
# Even though we set age limit for children as 16, it goes beyond that
# because it is subject to the bandwidth.
```

```python
fig = sns.FacetGrid(titanic_df, hue = 'Pclass', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()
fig.set(xlim = (0, oldest))

fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.factorplot('Cabin', data = cabin_df, kind = 'count', palette = 'winter_d')
```

```python
# more palettes can be found by searching for matplotlib colormaps
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
# removing the T value since it makes no sense

sns.factorplot('Cabin', data = cabin_df, palette = 'summer', kind = 'count')
```

```python
sns.factorplot('Embarked', data = titanic_df, hue = 'Pclass', kind = 'count', order = ['C', 'Q', 'S'])
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
# sibsp = siblings parch = parents/children

titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
# ignore warning
```

```python
titanic_df.head()
```

```python
sns.factorplot('Alone', data = titanic_df, palette = 'Blues', kind = 'count', order = ['Alone', 'With Family'])
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0: 'no', 1: 'yes'})
sns.factorplot('Survivor', data = titanic_df, palette = 'Set1', kind = 'count')
```

```python
sns.factorplot('Pclass', 'Survived', hue = 'person', data = titanic_df, order = [1, 2, 3])
```

```python
sns.lmplot('Age', 'Survived', data = titanic_df)
```

```python
# All the points at the top represent the people who survived and vice versa.
```

```python
sns.lmplot('Age', 'Survived', hue = 'Pclass', data = titanic_df, palette = 'winter')
```

```python
generations = [10, 20, 40, 60, 80]

sns.lmplot('Age', 'Survived', hue = 'Pclass', data = titanic_df, palette = 'winter', x_bins = generations)
```

```python
sns.lmplot('Age', 'Survived', hue = 'Sex', data = titanic_df, palette = 'winter', x_bins = generations)
```
